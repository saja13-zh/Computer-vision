```python
import pandas as pd
```

```python
apears = pd.read_csv('/kaggle/input/handwriting-data-set/train.csv')

y=apears.pop("label").values
x=apears.values
print(x)
print(y)
apears.shape
apears.head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 100)
print(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score

model2=KNeighborsClassifier(n_neighbors=3)
ap_tree=model2.fit(X_train, y_train)
predictions = ap_tree.predict(X_test)


accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy of KNeighborsClassifier with K=3: {accuracy:.2f}")
```
